```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.
```

```python
Matplotlib is a plotting library for the Python programming language that provides a wide variety of static, animated, and interactive visualizations in Python. Matplotlib is used for data visualization, exploration, and communication. It is used to create charts, graphs, and other types of visualizations that help in the interpretation of data.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

1. Line Plot: A line plot is a simple plotting technique to display data points in a series as a line segment. It is used to visualize the trend of data over time.

2. Scatter Plot: A scatter plot is used to visualize the relationship between two variables. Each point represents an observation and the position on the plot represents the values for the two variables.

3. Bar Plot: A bar plot is used to visualize categorical data, displaying the frequency or proportion of categories as bars. It is used to compare different groups or categories.

4. Histogram: A histogram is used to visualize the distribution of continuous data. It displays the frequency of data points in different bins or intervals.

5. Pie chart: A pie chart is used to display the proportion of different categories in a dataset. It is used for data with categorical variables and helps in understanding the contribution of each variable in the dataset.
```

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

```python
##A scatter plot is a type of plot used to display the relationship between two variables. It displays data points as a collection of dots on a 2D plane, where each dot represents an observation in the data. A scatter plot is useful in identifying patterns and trends in data.

##Here is the code to generate a scatter plot using the given data in Python:
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.title("Relationship between X and Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
ax[0,0].plot(x1, y1)
ax[0,0].set_title("Line Plot 1")

ax[0,1].plot(x2, y2)
ax[0,1].set_title("Line Plot 2")

ax[1,0].plot(x3, y3)
ax[1,0].set_title("Line Plot 3")

ax[1,1].plot(x4, y4)
ax[1,1].set_title("Line Plot 4")
for i in ax.flat:
    i.set(xlabel='X Axis', ylabel='Y Axis')
plt.show()

```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
#A bar plot is a graphical representation of data using rectangular bars. It is commonly used to represent and compare quantitative data associated with different categories. In a vertical bar plot, the height of each bar indicates the corresponding value of the data, while in a horizontal bar plot, the length of each bar represents the corresponding value.

#Bar plots are used to visually represent the distribution of data, to compare the values of different categories, and to show trends in the data.

import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.title("Company Profits")
plt.ylabel("Profit")
plt.show()
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit")
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
#A box plot is a method of graphical representation of distribution of data. It shows the distribution of data using quartiles and extreme values. The box represents the interquartile range (IQR) and the line inside the box represents the median. The whiskers extend from the box to the highest and lowest values within 1.5 times the IQR from the box. Any data points outside the whiskers are displayed as individual points. Box plots are used to identify outliers, to compare and contrast different data sets, and to show the symmetry and skewness in the data.

import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data)
plt.title("Box Plot")
plt.ylabel("Value")
plt.show()

#The `plt.boxplot()` function is used to plot the data on a box plot and the `data` variable is a list that contains the two data sets. The `title()` and `ylabel()` functions are used to add a title and y-axis label to the plot. The `show()` function is used to display the plot. The resulting plot will show the box and whisker plots of the two data sets side-by-side for easy comparison.
```
